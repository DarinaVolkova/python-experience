# Построение и исследование точки равновесия спроса/предложения

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
a = -1
b = 30
c = 36
d = 6
e = 20

#функция спроса
def D(q):
    return a*q + b

#функция предложения
def S(q):
    return ((q**2)/c) + (q/d) + e
```

```python
#построение графика для функций спроса и предложения
q = np.arange(0,20,0.01)
x  = a*q + b
y = ((q**2)/c) + (q/d) + e
plt.plot(q, x, color = 'r')
plt.plot(q, y, color = 'g')
plt.grid(True)
plt.show()

#нахождение равновесной цены
x0 = float(fsolve(lambda q: D(q) - S(q),1))
print('Равновесная цена = ', x0)
```

```python
 # колебания равновесной цены при условии, 
    #что коэф. заданы с ошибкой в пределах 10%
a += a*(0.1)
b += b*(0.1)
c += c*(0.1)
d += d*(0.1)
e += e*(0.1)
q = np.arange(0,20,0.01)
x  = a*q + b
y = ((q**2)/c) + (q/d) + e
plt.plot(q, x, color = 'r')
plt.plot(q, y, color = 'g')
plt.grid(True)
plt.show()
x0 = float(fsolve(lambda q: D(q) - S(q),1))
print('Равновесная цена = ', x0)
```

```python
#Метод простых итераций
def phi(x):
    return (60/7)-(1/42)*(x**2)
#phi(q)' = (-1/21)*q 
eps = 0.001
q0 = 1
q1 = phi(q0)
#Подсчет числа итераций
counter = 0
while abs(q1-q0) >= eps:
    q0 = q1
    q1 = phi(q0)
    counter +=1
print(q1, counter, sep ='\n')
```
